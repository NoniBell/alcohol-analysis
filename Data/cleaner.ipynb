```python
import pandas as pd
```

```python
#load in sales xlsx
sales_df = pd.read_excel('alcsales2017to2020.xlsx', sheet_name='Data')
sales_df
```

```python
#types of beverage: 1=spirits, 2=wine, 3=beer, 4=all beverages
#FIPS code: 2=Alaska, 8=Colorado, 12=Florida, 17=Illinois, 21=Kentucky, 25=Massachusetts, 27=Minnesota,
#29=Missouri, 38=North Dakota, 47=Tennessee

#convert type and FIPS to string
sales_df = sales_df.astype({'FIPS': 'object', 'Beverage': 'object'})
sales_df.dtypes
```

```python
#rename beverages
for index, row in sales_df.iterrows():
    if row.Beverage == 1:
        sales_df.at[index, 'Beverage'] = 'Spirits'
    elif row.Beverage == 2:
        sales_df.at[index, 'Beverage'] = 'Wine'
    elif row.Beverage == 3:
        sales_df.at[index, 'Beverage'] = 'Beer'
    elif row.Beverage == 4:
        sales_df.at[index, 'Beverage'] = 'All Beverages'

#rename FIPS to state names
for index, row in sales_df.iterrows():
    if row.FIPS == 2:
        sales_df.at[index, 'FIPS'] = 'Alaska'
    elif row.FIPS == 8:
        sales_df.at[index, 'FIPS'] = 'Colorado'
    elif row.FIPS == 12:
        sales_df.at[index, 'FIPS'] = 'Florida'
    elif row.FIPS == 17:
        sales_df.at[index, 'FIPS'] = 'Illinois'
    elif row.FIPS == 21:
        sales_df.at[index, 'FIPS'] = 'Kentucky'
    elif row.FIPS == 25:
        sales_df.at[index, 'FIPS'] = 'Massachusetts'
    elif row.FIPS == 27:
        sales_df.at[index, 'FIPS'] = 'Minnesota'
    elif row.FIPS == 29:
        sales_df.at[index, 'FIPS'] = 'Missouri'
    elif row.FIPS == 38:
        sales_df.at[index, 'FIPS'] = 'North Dakota'
    elif row.FIPS == 47:
        sales_df.at[index, 'FIPS'] = 'Tennessee'

sales_df
```

```python
#load in happiness xls
happiness_df = pd.read_excel('worldhappinessdata.xls')
happiness_df

#capitalize and rename columns
happiness_df = happiness_df.rename(columns = {'year': 'Year', 'Positive affect': 'Positive Effect', 'Negative affect': 'Negative Effect'})

happiness_df
```

```python
spirits_df = pd.read_csv('spirits-consumption-per-person.csv')

#rename fourth column
spirits_df = spirits_df.rename(columns = {'Indicator:Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol) - Beverage Types:Spirits': 'Spirits Consumed (Liters)'})

wine_df = pd.read_csv('wine-consumption-per-person.csv')

#rename fourth column
wine_df = wine_df.rename(columns = {'Indicator:Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol) - Beverage Types:Wine': 'Wine Consumed (Liters)'})

beer_df = pd.read_csv('beer-consumption-per-person.csv')

#rename the fourth column
beer_df = beer_df.rename(columns = {'Indicator:Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol) - Beverage Types:Beer': 'Beer Consumed (Liters)'})
```

```python
#send cleaned dataframes to new csvs
happiness_df.to_csv('cleaned_happiness.csv', index=False)
sales_df.to_csv('cleaned_sales.csv', index=False)
```

```python
spirits_df.to_csv('cleaned_spirits.csv', index=False)
wine_df.to_csv('cleaned_wines.csv', index=False)
beer_df.to_csv('cleaned_beers.csv', index=False)
```
